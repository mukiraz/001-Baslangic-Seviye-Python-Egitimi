# Stack (yığın) kavramı

Yığın, programlamada yararlı bir veri yapısıdır. Tıpkı üst üste yerleştirilmiş cd yığını gibidir.

Yukarıdaki gibi bir cd yığınına yeni bir cd eklemek istersek en üste eklemek zorundayız. Bu cd yığınından bir cd almak istersek de en üsttekini alacağız. 
Eğer biz en alttaki cd'ye olaşmak istiyorsak, en alttaki ile beraber tüm cd'leri çıkarıp, en alttakini alıp sonra kalan hepsini geri takmak zorundayız. İşte böyle bir yapıya **Son Giren İlk Çıkar** yani LIFO (Last in first out) deriz.

LIFO mantığı ile çalışan veri setlerine programcılıkta **yığın** yani **stack** denir.

Yukarıdaki resimde, 3. veri en son eklenmiştir. Fakat, önce kaldırılmıştır. Bu nedenle , Son Giren İlk Çıkar (LIFO) ilkesini izler .

Programlamada, yığının en üstüne bir öğe koymaya **itme (push)**, bir öğeyi kaldırmaya ise **pop (çıkma)** denir.
