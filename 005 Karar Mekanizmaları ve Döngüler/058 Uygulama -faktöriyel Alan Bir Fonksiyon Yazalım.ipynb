# Faktöriyel Alan Fonksiyon

Bir fonksiyon oluşturalım ve bu fonksiyon tek parametre alsın. Argüman olarak bir tamsayı gönderdiğimizde, bize bu sayının faktöriyelini hesaplayıp ekrana yazdırsın.

```python
# factorial function consists of a single parameter. If the argument is 0 or 1, the result is 1. for integers greater than 1
# we multiply each value we create by subtracting 1 from this number.
# If our number is "5", the factorial will be 5 x 4 x 3 x 2 x 1 = 120
def factorial(number):
    result = 1
    if (number == 0 or number == 1):
        print("{} sayıının faktöriyeli {}".format(number,result))
    else:
        while (number >= 1):
            result *= number
            number -= 1
        print("{} sayıının faktöriyeli {}".format(number,result))
```

```python
factorial(0)
```

```python
factorial(5)
```
