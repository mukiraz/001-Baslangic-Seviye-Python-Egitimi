# Modüller ve import kullanımı

Python yorumlayıcısından çıkıp tekrar girerseniz, yaptığınız tanımlar (işlevler ve değişkenler) kaybolur. Bu nedenle, biraz daha uzun bir program yazmak istiyorsanız, yorumlayıcı için girdiyi hazırlamak için bir metin düzenleyici kullanmak ve bunun yerine girdi olarak bu dosya ile çalıştırmak daha iyidir. Bu, komut dosyası oluşturmak olarak bilinir . Programınız uzadıkça, daha kolay bakım için birkaç dosyaya bölmek isteyebilirsiniz. Ayrıca, tanımını her programa kopyalamadan birkaç programda yazdığınız kullanışlı bir işlevi kullanmak isteyebilirsiniz.

Bunu desteklemek için Python'un tanımları bir dosyaya koymanın ve bunları bir komut dosyasında veya yorumlayıcının etkileşimli bir örneğinde kullanmanın bir yolu vardır. Böyle bir dosyaya modül adı verilir ; bir modüldeki tanımlar diğer modüllere veya ana modüle içe aktarılabilir .

Modül, Python tanımlarını ve ifadelerini içeren bir dosyadır. Dosya adı, sonu .py ile birlikte modül adıdır. Bir modül içinde, modülün adı (bir dizge olarak) _name_ global değişkenin değeri olarak mevcuttur.

Pythonda her bir dosya aslında bir **modüldür** ve her bir modül içinde fonksiyonlar, sınıflar ve objeler barındırır. Biz de bu modülleri programımıza **dahil ederek** içindeki fonksiyonlardan, sınıflardan ve objelerden faydalanabiliriz.

Aslında şu ana kada pythonda gömülü olan (built-in) bir çok modül kullandık. Matematiksel hesaplamaları yaparken kullandığımız math modülü gibi. Bu modüller, aslında birer .py dosyalarıdır. Python programının kurulu olduğu dosyaları görebilmek için aşağıdaki komutları girebiliriz.

```python
import os
import sys
os.path.dirname(sys.executable)
```

```python
import sys
sys.executable
```

Python un kurulu olduğu klasörde, "Lib" klasörünün içinde, pythonda yazılmış gömülü (built-in) modülleri görebiliz.

Veya aşağıdaki kodla built-in fonksiyonlara erişebiliriz.;

```python
import sys
sys.builtin_module_names
```

Burada, "Lib" klasörünün içerisinde bulunan kütüphanelerle, sys.builtin_module_names komutunun çıktısı olan modüller farklılık göstermektedir. Örneğin daha önce de kullandığımız math modülü "lib" klasörünün altında yer almamaktadır. Bunun nedeni, math ve sys modüllerinin hızlı çalışması amaçlanmıştır. Bu maksatla bu modüller C ile yazılmış  ve doğrudan Python yorumlayıcısına dahil edilmiştir.

Pythonda da modül kavramı oldukça önemli bir kavramdır. Bir Python modülünü programımıza dahil ederek bu modüller içinde yazılan fonksiyonlardan  ve sınıflardan kullanabilir ve programlarımızı daha efektif bir şekilde yazabiliriz. Eğer modül diye bir kavram olmasaydı, programlarımızdaki herbir fonksiyonu ve sınıfı kendimiz yazmak zorunda kalacaktık.

Pythonda Python geliştiricileri tarafından yazılmış bir çok hazır modül bulunmaktadır. Ayrıca , programcıların internete ve **Githuba** yükledikleri bir çok modülü programlarımızda kullanabilir ve daha güzel programlar yazabiliriz.
