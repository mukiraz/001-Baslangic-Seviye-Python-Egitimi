- **Ücretsiz ve açık kaynaklı:** 
    Anaconda ücretsiz ve açık kaynaklıdır. Bu, veri bilimcileri ve veri analistlerinin maliyetinden tasarruf etmelerini sağlar.

**Anacondanın sağladığı bazı spesifik faydalar şunlardır:**

## Anaconda İle Hangi IDE'leri İçerir?

Anaconda, Python için bir dağıtımdır. Bu, Python'un yanı sıra popüler veri bilimi ve makine öğrenimi kütüphanelerini ve araçlarını da içerir. Anaconda ile gelen IDE'ler şunlardır:

* **Spyder:** Spyder, Python için ücretsiz ve açık kaynaklı bir IDE'dir. Veri bilimi ve makine öğrenimi için özel olarak tasarlanmıştır. Spyder, kod düzenleme, hata ayıklama, grafik oluşturma ve veri analizi için çeşitli araçlar sunar.
* **Jupyter Notebook:** Jupyter Notebook, etkileşimli bir ortam sağlayan bir IDE'dir. Kod, metin, Markdown, ve resim gibi farklı içerik türlerini aynı defterde birleştirebilirsiniz. Jupyter Notebook, veri bilimi ve bilimsel hesaplama için popüler bir araçtır.
* **PyCharm:** PyCharm, Python için ücretli bir IDE'dir. Profesyonel geliştiriciler için tasarlanmıştır. PyCharm, kod düzenleme, hata ayıklama, grafik oluşturma ve veri analizi için çeşitli araçlar sunar.

Anaconda ayrıca, Python için bir dizi başka IDE'yi de destekler. Bunlar şunlardır:

* **IDLE:** IDLE, Python için ücretsiz ve açık kaynaklı bir IDE'dir. Temel kod düzenleme ve hata ayıklama için tasarlanmıştır.
* **Visual Studio Code:** Visual Studio Code, Microsoft tarafından geliştirilen ücretsiz ve açık kaynaklı bir IDE'dir. Python için birçok eklenti mevcuttur.
* **Sublime Text:** Sublime Text, ücretli bir metin editörüdür. Python için birçok eklenti mevcuttur.

Anaconda'yı kurarken, hangi IDE'lerin yükleneceğini seçebilirsiniz.



## Neden Derslerimizde Jupyter'i Kullanacağız?

Jupyter Notebook, veri bilimi ve bilimsel hesaplama için etkileşimli bir ortam sağlayan açık kaynak kodlu bir programdır. Python, R, Julia, ve diğer birçok programlama dilini destekler.

Jupyter Notebook, bir defter gibi çalışır. Her defter sayfası bir veya daha fazla hücreden oluşur. Hücreler kod, metin, Markdown, ve resim içerebilir. Kod hücreleri çalıştırıldığında, sonuçlar hücrenin altına yazdırılır.

Jupyter Notebook, veri bilimi ve bilimsel hesaplama için popüler bir araçtır çünkü:

* **Etkileşimlidir:** Jupyter Notebook, kod hücrelerini tek tek çalıştırmanıza ve sonuçları hemen görmenize olanak tanır. Bu, araştırma ve geliştirme için idealdir.
* **İçeriği zengindir:** Jupyter Notebook, kod, metin, Markdown, ve resim gibi farklı içerik türlerini aynı defterde birleştirebilirsiniz. Bu, veri bilimi projeleriniz için dokümantasyon ve açıklamalar oluşturmayı kolaylaştırır.
* **Paylaşılabilir:** Jupyter Notebook defterlerinizi diğerleriyle kolayca paylaşabilirsiniz. Bu, işbirliği ve tekrarlanabilirlik için idealdir.

Jupyter Notebook, veri bilimcileri, veri analistleri, araştırmacılar, ve diğer profesyoneller tarafından yaygın olarak kullanılmaktadır.

## Jupyter Notebook'u Çalıştırmak

Anaconda Python Dağıtımını kurduktan sonra, Jupyter Notebook'u windows programlarından bulup çalıştırın.

Jupyter web browser üzerinde çalışan nir uygulamadır. Jupyter'i çalıştırdıktan hemen sonra bir komut ekranı karşınıza gelir. Bu komut ekranı aşağıdaki gibidir. 

Anaconda yazılımını kurarken aşağıdaki seçeneği seçtiyseniz eğer bu sizin bilgisayarınızdaki kullanıcı klasörünü gösterecektir. Jupyter doğrudan burada açılır.



Komut satırı çalıştıktan sonra varsayılan web tarayıcınız açılır. Eğer açılmazsa, web tarayıcınızın adres satırına, yukarıdaki resimde kırmızı bölümde yer alan
- http://localhost:8888/?token=token
veya
- http://127.0.0.1:8888?token=token

adresini kopyalayıp yapıştıtabilirsiniz. Jupyter her başladığında yeni bir token oluşturur. Bu token güvenlik içindir. Bu token olmazsa, jupyter ugulaması web browser'da görülmeyecektir.

Jupyter uygulaması aşağıdaki gibi başlayacaktır.



### Menü açıklamaları

1. Klasörleri gösterir. Klasörler arasında dolaşabilirsiniz.
2. Açık olan jupyter dosyalarını listeler.
3. Clusters sekmesi artık IPython paralel tarafından sağlanmaktadır. 
4. Bir jupyter dosyasını kaydetmek için kullanılır.
5. Yeni 

    * **Python 3:** Yeni bir jupyter dosyası oluşturur.
    * **Text file:** Yeni bir not defteri oluşturur
    * **Folder:** Yeni bir klasör oluşturur.
    * **Terminal:** Yeni bir cmd komut satırı uygulaması açar.
6. Sayfayı yeniler.
7. Jupyter programını sonlandırır.
8. Oturumu sonlandırır. Yeniden bağlanmak için jupyter ilk çalıştığında çıkan komu satırındaki token'i girmelisiniz!

### Yeni bir Jupyter dosyası oluşturmak

Jupyter notebook açıldıktan sonra "new" sekmesine basınız ve Python3 sekmesini seçerek yeni bir dosya oluşturun. Bu dosya sizin içerisinde olduğunuz klasörde oluşturulacaktır. Dosya oluşturulduktan sonra boş bir jupyter dosyası aşağıdaki gibi oluşturulmuş olacaktır.




1. Dosya adı. Dosya adı burada yer alır. Buraya tıklayarak dosya adını değiştirebilirsiniz.
2. Oturum sonlandırma işlemi.
3. Dosya menüsü.
4. Düzenle menüsü.
5. Görünüm menüsü.
6. Ekle menüsü.
7. Hücre menüsü.
8. Kernel (Çekirdek) menüsü.
9. Araçlar menüsü.
10. Yardım menüsü.
11. Kaydet
12. Hücre ekle
13. Kes
14. Kopyala
15. Yapıştır
16. Hücreyi yukarı al
17. Hücreyi aşağı al
18. Python kodunu yürüt.
19. Python kodunu durdur.
20. Kernel (Çekirdek)'ı sıfırla. (Bu düğmeye basıldığında tüm değişkenler ve objeler RAM'den silinir.)
21. Kernel'ı sıfırla ve tüm hücreleri yeniden çalıştır.
23. Hücre İçeriğini Seçme:
    * **Code:** Python kodu yazıp çalıştırabilirsiniz.
    * **Markdown:** Markdown, düz metin kullanarak zengin metin biçimlendirmesi oluşturmak için kullanılan bir işaretleme dilidir. Markdown, HTML'ye kolayca dönüştürülebilir, bu da onu web siteleri, bloglar, belgeler ve diğer çevrimiçi ortamlarda kullanım için popüler kılar. Bu sayede Markdown language ile metinlerinizi yazabilirsiniz. Daha Fazla Bilgi için:
        - https://docs.github.com/en/get-started/writing-on-github/getting-started-with-writing-and-formatting-on-github
    * **Raw NBContent:** Jupyter'daki Raw NBContent, not defterinize herhangi bir şekilde işlenmeden veya dönüştürülmeden nbconvert çıktısına geçirilecek ham metin eklemenizi sağlayan bir hücre türüdür
    * **Heading:** Markdown heading
24. Komut Paleti: Buradan tüm menü komutlarına ulaşabilirsiniz.
