# Ondalıklı Sayılar (Float)

Ondalıklı sayılar da bizim matematikte gördüğümüz sayı çeşitlerinden bir tanesidir. Ondalıklı sayılar Pythonda da bir veri tipidir. Ondalıklı Sayılar veya diğer adıyla Kayan Sayılar İngilizce olarak **Float** olarak geçmektedir.

### Ondalıklı Sayılar Bilgisayarda Nasıl Tanımlanır?

İkili sayı sisteminde ondalıklı sayılar, 0 ve 1 rakamları kullanılarak ifade edilir. İkili sayı sistemi, 2'ye tabanlı bir sayı sistemidir. Bu, her basamağın 2'nin bir kuvvetini temsil ettiği anlamına gelir.

İkili sayı sisteminde ondalıklı sayılar, aşağıdaki şekilde ifade edilebilir:

* **Kesirli ondalıklı sayılar:** Kesirli ondalıklı sayılar, 2'nin bir kuvveti olan tabanla çarpılarak ifade edilir. Örneğin, 1011.011 sayısı, 2^3 + 2^2 + 2^0 + 2^-1 + 2^-2 = 13.09375 sayısını temsil eder.
* **Tam ondalıklı sayılar:** Tam ondalıklı sayılar, 2'nin bir kuvveti olan tabanla çarpılarak ifade edilir. Örneğin, 10110 sayısı, 2^3 + 2^2 + 2^0 = 13 sayısını temsil eder.

İkili sayı sisteminde ondalıklı sayılar, aşağıdaki şekilde dönüştürülebilir:

* **Ondalıklı sayıdan ikili sayıya:** Ondalıklı sayı, 2'nin bir kuvveti olan tabanla çarpılarak ve kalanı kaydedilerek dönüştürülür. Kalan 0 olana kadar bu işlem tekrarlanır. Kalanlar, ikili sayının basamaklarını oluşturur. Örneğin, 13.09375 sayısının ikili karşılığı, 1011.011'dir.
* **İkili sayıdan ondalık sayıya:** İkili sayı, en sağdaki basamaktan en soldaki basamağa doğru, basamakların değerlerini sırasıyla çarpılarak dönüştürülür. Sonuç, ondalıklı sayıyı temsil eder. Örneğin, 1011.011 sayısının ondalık karşılığı, 13.09375'tir.

İkili sayı sisteminde ondalıklı sayılar, bilgisayarlar tarafından kullanılan sayı sistemidir. Bu nedenle, bilgisayar programlamada ondalıklı sayılar genellikle ikili sayı sistemi kullanılarak ifade edilir.

İşte ikili sayı sisteminde ondalıklı sayıların ifade edilmesine ilişkin bazı örnekler:

* **Kesirli ondalıklı sayılar:**

    * 1011.011: 13.09375
    * 11001.101: 25.59375
    * 10000000.0001: 2^10 + 2^-10 = 1024.0001

* **Tam ondalıklı sayılar:**

    * 10110: 13
    * 110010: 25
    * 10000000: 2^10 = 1024

İkili sayı sisteminde ondalıklı sayıların ifade edilmesi, bilgisayar programlamada ondalıklı sayılarla çalışmak için önemlidir.

### Ondalıklı Sayılar RAM'de nasıl tanımlanır?

Python programlama dilinde, float sayılar, Python'ın ondalıklı sayılar için kullandığı veri türü olan float ile tanımlanır. Float sayılar, ondalıklı bir sayı olarak ifade edilir. Örneğin, 3.14, 3.554546, -13.54 gibi sayılar Python'da birer float değeridir.

Python'da float sayılar, RAM'de bir dizi bit olarak depolanır. Her bit, bir 0 veya 1 değerini alabilir. Float sayının işareti, en soldaki bit tarafından saklanır. 0, pozitif bir sayı anlamına gelirken, 1 negatif bir sayı anlamına gelir.

Float sayıların ikili sayı sistemine çevrilebilir. Matematiksel hesaplamaları gerektirir.
