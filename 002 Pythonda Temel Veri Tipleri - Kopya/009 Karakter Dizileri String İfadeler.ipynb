Stringleri bizler ""(tırnak) işaretleri ile oluştururuz.

```python
"Murat Ugur KIRAZ"
```

```python
'Murat Ugur KIRAZ'
```

#### Üç tırnakla Oluşturma

```python
"""Murat Ugur KIRAZ"""
```

### String ifadeler Bilgisayarda nasıl tanımlanır?

String ifadeler, RAM'de bir dizi karakter olarak tutulur. Her karakter, bir ASCII veya Unicode karakterinin kodu olarak temsil edilir. ASCII, 8 bitlik bir kodlama sistemidir. Unicode ise, 16 bitlik veya 32 bitlik bir kodlama sistemidir.

ASCII Kod tablosu aşağıda yer almaktadır:





Örneğin:

```python
s = "Hello"
```

Şeklinde bir string tanımladığımızda, s değişkeni bir string objesidir. Bilgisayar dilinde bir obje, RAM'de yer işgal eden ve içerisinde metotları (fonksiyonları), verileri barındıran bir veri yapısıdır. String bir değişkenin içerisinde, veri olarak, her bir karakterin matematiksel değeri vardır.

s = "Hello" stringinde, her bir harf ASCII veya Unicode kod tablosunda belirtilen sayıya denk gelir ve RAM'de tutulur.



Python C tabanlı bir dildir. C dilinde string ifadeler, karakter dizileri şeklinde şekilde görüldüğü gibi tanımlanmıştır. C dilinde her string ifadenin sonunda null yani 00h değeri yer alır. 

s değişkeninde aslında string ifadenin ilk karakterinin RAM'deki adresi tutulur. Diğer karakterler yan yana sıralanmıştır. s değişkenini çağırdığımızda, bilgisayar string ifadenin ilk adresinden okumaya başlar, null ifadesini görene kadar okumaya devam eder. Yani bilgisayar string bir ifadenin bittiğini 8 bitlik sıfırla algılar.

Bu nedenle string ifadelerin herhangi bir karakterini değiştiremeyiz. Değiştirmek istersek de, ilgili stringi RAM'de başka bir alana kopyalamamız gerekir. String işlemleri maliyetlidir.

Burada s değişkeni aslında bir string objesidir. İçerisinde metotları barındırır.

s. dedikten sonra TAB tuşuna basarsak, s objesine ait metotları görebiliriz.




```python
s[2] = "f" # Bu kod hata verir. Bu kod ile s değişkeninin soldan 3'üncü karakteri "f" yapılmak istenmiştir.
```
