# Sıcaklık Birimleri Dönüşümleri

Şimdi sıcaklık birimlerini dönüştüren basit bir program yazalım.

Yukarıdaki ifadeyi kullanarak sıcaklık birimleri arasındaki dönüşümleri yapabiliriz. Bunun için gerekli kodlamaları aşağıdaki şekilde yazabiliriz.

```python
# Önce print fonksiyonu ile kullanıcılarımızı bilgilerndirelim.
print("""-------------------------------------------------------
Bu program, Celcius derece cinsinden girilen değeri Fahrenheit'a çevirir.
-------------------------------------------------------
""")
         
celcius = float(input("Lütfen celcius cinsinden bir değer giriniz:"))
fahrenheit = ((celcius*180)/100)+32
mesaj = "{} celcius derecenin fahrenheit karşılığı {} fahrenheit'dır".format(celcius,fahrenheit)
print(mesaj)     
```

```python
# Önce print fonksiyonu ile kullanıcılarımızı bilgilerndirelim.
print("""-------------------------------------------------------
Bu program, Fahrenheit cinsinden girilen değeri Celcius'a çevirir.
-------------------------------------------------------
""")
         
fahrenheit = float(input("Lütfen celcius cinsinden bir değer giriniz:"))
celcius = ((fahrenheit-32)/180)*100
mesaj = "{} fahrenheit'ın celcius karşılığı {} celcius'tur".format(fahrenheit,celcius)
print(mesaj) 
```

```python
# Önce print fonksiyonu ile kullanıcılarımızı bilgilerndirelim.
print("""-------------------------------------------------------
Bu program, Celcius derece cinsinden girilen değeri Kelvin'e çevirir.
-------------------------------------------------------
""")
         
celcius = float(input("Lütfen celcius cinsinden bir değer giriniz:"))
kelvin = celcius+273
mesaj = "{} celcius derecenin Kelvin karşılığı {} Kelvin'dir".format(celcius,kelvin)
print(mesaj)
```

```python
# Önce print fonksiyonu ile kullanıcılarımızı bilgilerndirelim.
print("""-------------------------------------------------------
Bu program, Kelvin cinsinden girilen değeri Celcius'a çevirir.
-------------------------------------------------------
""")
         
kelvin = float(input("Lütfen kelvin cinsinden bir değer giriniz:"))
celcius = kelvin-273
mesaj = "{} Kelvin'in celcius karşılığı {} celcius'tur".format(kelvin,celcius)
print(mesaj)
```
