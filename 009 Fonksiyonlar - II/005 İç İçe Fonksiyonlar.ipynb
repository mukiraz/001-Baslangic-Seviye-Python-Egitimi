# İç İçe Fonksiyonlar

Python'daki fonksiyonlar birinci sınıf nesnelerdir. Herhangi bir kapsamda tanımlanabilirler.

```python
def fibonacci(n):
    def digit(a,b):
        return b, a + b
    a, b = 0, 1
    list=[]
    for i in range(n):
        a, b = digit(a, b)
        list.append(b)
    return list

print(fibonacci(20))
```

Fonksiyonlar, kapsama alanlarını yakalar ve diğer herhangi bir nesne türü gibi geçirilebilir

```python
def get_sum(n):
    def adder(x):
        return n + x
    return adder

add5 = get_sum(5)
add6 = get_sum(6)
print(add5(10)) #Çıktı: 15

print(add6(10)) #Çıktı: 16
```

```python
def add_repetitive(func, k, x):
    for i in range(k):
        x = func(x)
    return x
```

```python
print(add_repetitive(add5, 5, 1)) #Çıktı: 26
```
