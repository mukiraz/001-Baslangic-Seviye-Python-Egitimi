# Atanmış Değişkenleri Kullanarak Rekürsif Lambda Oluşturmak

Özyinelemeli lambda fonksiyonları oluşturmanın bir yöntemi, işlevin bir değişkene atanmasını ve ardından bu değişkene işlevin içinde başvurulmasını içerir. Bunun yaygın bir örneği, aşağıdaki kodda gösterildiği gibi, bir sayının faktöriyelinin recursive hesaplamasıdır:

```python
lambda_func = lambda i : 1 if i == 0 else i * lambda_func(i-1)
print(lambda_func(4)) # 4 * 3 * 2 * 1 = 12 * 2 = 24
```

Lambda işlevi, değişken ataması yoluyla, 0 ise değerlendirip 1 döndürdüğü bir değer (4) geçirilir veya aksi takdirde, -1 (i-1) değerinin lambda işlevi tarafından başka bir hesaplama i* geçerli değerini döndürür. Bu, aktarılan değer 0'a düşene kadar devam eder (dönüş 1). Şu şekilde görselleştirilebilen bir süreçtir.
