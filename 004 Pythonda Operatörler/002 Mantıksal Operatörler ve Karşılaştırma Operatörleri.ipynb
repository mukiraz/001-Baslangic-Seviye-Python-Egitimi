# Mantıksal Operatörler

Bilgisayarların 1'ler ve 0'lardan oluştuğunu söylemiştik. Bilgisayar temel anlamda 1 ve 0 lardan oluşan yapıları yöneten bir sistemler bütünüdür de diyebiliriz. Mantıksal operatörleri anlayabilmemiz için temel bilgisayar mantık kapılarını anlamamız gerekmektedir.

Not: Yukarıdaki görsel http://javigcsecs.blogspot.com/2017/04/131-logic-gates.html adresinden alınmıştır.

Aslında, 1 sayısını doğru (true), 0 sayısını da yanlış (false) olarak düşünebiliriz.

## *ve (and)* mantık kapısı

Bu mantıksal bağlaç, bütün karşılaştırma işlemlerinin sonucunun True olmasına bakar. Bağlanan karşılaştırma işlemlerinin hepsinin kendi içinde sonucu True ise genel sonuç True , diğer durumlarda ise sonuç False çıkar. Kullanımı şu şekildedir

```python
a=False
b=False

print(a and b)

a=True
b=False

print(a and b)

a=False
b=True

print(a and b)

a=True
b=True

print(a and b)
```

## *veya (or)* mantık kapısı

Bu mantıksal bağlaç, bütün karşılaştırma işlemlerinin sonuçlarından **en az birinin True** olmasına bakar. Bağlanan karşılaştırma işlemlerinin **en az birinin True** olmasında genel sonuç **True** , diğer durumlarda ise sonuç **False** çıkar. Kullanımı şu şekildedir.

```python
a=False
b=False

print(a or b)

a=True
b=False

print(a or b)

a=False
b=True

print(a or b)

a=True
b=True

print(a or b)
```

## *değil (not)* mantık kapısı

*not*  operatörü aslında bir mantıksal bağlaç değildir. Bu operatör sadece bir mantıksal değeri veya karşılaştırma işlemininin tam tersi sonuca çevirir. Yani, not operatörü **True** olan bir sonucu **False** , **False** olan bir sonucu **True** sonucuna çevirir. Kullanımı şu şekildedir.

```python
a=False

print(a)
print(not a)

```

## is operatörü

İki nesnenin aynı nesne olup olmadığını kontrol eder. Sonuç Boolean (True) (False) döner. is İki değişken aynı nesneyi gösteriyorsa kelime testi için kullanılır. 

İki nesne aynı nesneyse True döndürür.

İki nesne aynı değerleri içerse bile farklı nesneler ise sonuç False döner.

Bu yüzden iki değişkenin eşit olup olmadığını test etmek için == operatörünü kullanın.

```python
x = ["elma", "muz"]
y = ["elma", "muz"]
z = x

print(x is z)

# True döner çünkü z ve x aynı nesnedir.

print(x is y)

# x ve y aynı değere sahip olsalar bile False döner çünkü aynı obje değillerdir.

print(x == y)

# True döner çünkü x ve y aynı değere sahiptir.
```

### Karşılaştırma operatörlerinin sonuçlarını mantıksal operatörlerle karşılaştırma

Karşılaştırma operatörlerinin sonuçlarını mantıksal operatörlerle karşılaştırabiliriz.

Örneğin; 
**5 sayısı hem 4 sayısından hem de 6 sayısından büyük mü?**
İki şartında da sağlanması bekleniyor. Fakat 5 sayısı 6' dan küçük olduğu için ikinci şartı sağlamıyor. Yani ifadenin sonucu *YANLIŞ* olacaktır.

Ya da 

**5 sayısı 4 sayısından veya 6 sayısından büyük mü?**
Bu örnekteki bekletide ise iki şarttan sadece bir tanesinin doğru olması yeterli. Bu sebeple ifadecenin sonucu *DOĞRU* olacaktır.

Uykarıdaki iki örnek için python kodlarını yazalım;

```python
a = 5
b = 4
c = 6

# 5 sayısı hem 4 sayısından hem de 6 sayısından büyük mü?
print((a > b) and (a > c))

# 5 sayısı 4 sayısından veya 6 sayısından büyük mü?

print((a > b) or (a > c))
```
