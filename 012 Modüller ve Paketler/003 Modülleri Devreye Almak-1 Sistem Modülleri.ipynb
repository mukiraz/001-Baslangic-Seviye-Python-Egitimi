# Modülleri Devreye Almak

## Sistem (Built-in) Modülleri Devreye Almak

İlk derslerimizde, math modülünü devreye almıştık. Bunu aşağıdaki gibi gerçekleştirmiştik.

```python
import math
```

Bir modülün içeriğinde ne olduğunu öğrenmek istiyorsak, dir() fonksiyonunu kullanırız. dir() fonksiyonu,  bir modülün hangi isimlerin (names) tanımladığını bulmak için kullanılır. Sıralı bir dizge listesi döndürür:

```python
dir(math)
```

Modüllerin görevlerini görmek için help() fonksiyonunu kullanırız.

```python
help(math)
```

Biz bir modülün ilgili fonksiyonunu kullanmak istiyorsak, math.fonksiyonadi() şeklinde kullanmamız gerekir.

```python
math.factorial(5)
```

Aynı şekilde help() fonksiyonu ile de modüllerin içeriğindeki fonksiyonların ne yaptığını görebiliriz.

```python
help(math.factorial)
```

### Modülü kendi belirlediğimiz isimle kullanmak

```python
import math as matematik
```

```python
matematik.factorial(5)
```
